```python
# Cleaning Data
import pandas as pd

# Missing values
df = pd.read_csv("raw_data.csv")

df.isnull()   #returns true if empty
# df.isna()  #same as isnull()

df.isnull().sum()

df.dropna() #drops/removes empty cell rows
# df.dropna(axis=1) #column drop

df.fillna(0)

age_mean = df["age"].mean()

cleaned_data = df.copy()
age_mean = cleaned_data["age"].mean()
cleaned_data["age"] = cleaned_data["age"].fillna(age_mean)

df.ffill()  # forward fill
df.bfill() # backward fill
```

```python
# Handle Duplicates
df.duplicated()

df.drop_duplicates()
```

```python
# Data Types

df.dtypes
df2 = df.copy()

df2 = df2.fillna(0)
df2["age"] = df2["age"].astype("int64").copy()
df2.dtypes

#to_datetime
date_str = pd.Series( [pd.to_datetime("4-12-2025")] )
type(date_str.dtypes)
```
