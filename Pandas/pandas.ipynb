```python
import pandas as pd
```

```python
# Series => 1D labelled array -> individual DataFrame column
s = pd.Series([1, 2, 3, 4, 5])
print(s)
print(type(s))

#index
print(s[2])
print(s[4])
```

```python
#Custom labels using index attribute
s = pd.Series([10, 23, 31, 45, 52], index = ["Hima", "Anil", "Subbu", "jammu", "Gouse"])

print(s)
print(s["jammu"])
print(s["Subbu"])
print(s.index)
```

```python
# Series Properties
s1 = pd.Series([1, 2, 3, 4, 5])
s2 = pd.Series([10, 23, 31, 45, 52])

print(s1 + s2)

#mutable values but immutable size
s1[1] = 20
changed_s1 = s1.drop(1)

print(s1)
print(changed_s1)
```

```python
# DataFrame => 2D labelled array

# Dictionary of list
info = {
    "Name": ["Hima", "Anil", "Subbu"],
    "Age": [22, 23, 24],
    "CGPA": [8.5, 8.6, 9.1]
}

df = pd.DataFrame(info)
print(df, type(df))
print(df.index)
print(df.columns)
```

```python
# list of list
df = pd.DataFrame([["Hima", 22], ["Anil", 23], ["Subbu", 24]], columns=["Name", "Age"] )

print(df)
```

```python
# Using numpy arrays
import numpy as np

np_arr = np.array([[1, 2, 3], [4, 5, 6]])

df = pd.DataFrame(np_arr, columns=["A", "B", "C"])

print(df)
```
