```python
import pandas as pd
```

```python
# Reading the data
data = pd.read_csv("IRIS.csv")
```

```python
# Summary Statistics
data.describe()
```

```python
# B. Selecting the rows based on conditions
# petal_length > 4.5 and species = Iris_virginica

data.query("petal_length > 4.5 & species == 'Iris-virginica'")
```

```python
# C. Group by species

data.groupby("species").agg({
    "sepal_length": "mean",
    "petal_width": "max",
    "sepal_width": "std"
})

data.groupby("species").agg(
    avg_sepalLength = ("sepal_length", "mean"),
    max_petalWidth = ("petal_width", "max"),
    std_sepalWidth = ("sepal_width", "std")
)
```

```python
# D. Create a new column petal_ratio

data["petal_ratio"] = data["petal_length"] / data["petal_width"]
data.tail(10)
```
