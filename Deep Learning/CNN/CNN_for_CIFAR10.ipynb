```python
import torch
import torch.nn as nn
import torch.optim as optim

import torchvision
from torchvision.datasets import CIFAR10
```

```python
# Datasets & DataLoaders
from  torch.utils.data import DataLoader
import torchvision.transforms as transforms

# image => scale(0, 1) => normalize(-1, 1)
transform = transforms.Compose([
    transforms.ToTensor(), # it transform dataset into Tensor and automatically scale it to (0, 1)
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
])

trainset = CIFAR10(root="./data", train=True, download=True, transform=transform)
testset = CIFAR10(root="./data", train=False, download=True, transform=transform)
```

```python
trainloader = DataLoader(trainset, batch_size=64, shuffle=True)
testloader = DataLoader(testset, batch_size=64)
```

### Build the CNN

```python
class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()

        self.conv_layers = nn.sequential(
            nn.convo2D(3, 32, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2D(2, 2), # kernel_size = 2 & stride = 2

            nn.convo2D(32, 64, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2D(2, 2),

            nn.convo2D(64, 128, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2D(2, 2)
        )
```
