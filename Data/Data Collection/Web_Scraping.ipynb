```python
# WEB SCRAPING

import requests
```

```python
URL = "https://www.scrapethissite.com/pages/simple/"

res = requests.get(URL)
```

```python
if res.status_code == 200:
    print(res.headers)
else:
    print("Failed to fetch the data")

with open("Scraped _data/data1.html", "w", encoding="utf-8") as f:
    f.write(res.text)
```

```python
# Beautiful Soup

from bs4 import BeautifulSoup

with open("Scraped _data/data1.html", "r", encoding="utf-8") as f:
    html_content = f.read()

soup = BeautifulSoup(html_content, "lxml")
```

```python
all_h3 = soup.find_all("h3")

all_countries = []

for h3 in all_h3:
    name = h3.get_text(strip=True)

    population = h3.find_next("div").select("span.country-population")[0].get_text(strip=True)

    all_countries.append([name, population])
```

```python
import pandas as pd

df = pd.DataFrame(all_countries, columns=["Name", "Population"])
```

```python
df.to_csv("Scraped _data/data.csv", index=False)
```
