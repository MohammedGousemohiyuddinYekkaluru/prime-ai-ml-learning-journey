```python
import json
```

```python
# Loading the data
def load_data(filename):
    with open(filename, "r") as f:
        data = json.load(f)

    return data    
```

```python
data = load_data("store_data.json")
```

```python
# Cleaning and structuring the data

def cleaning_data(data):
    text_to_num = {"one": 1, "two": 2, "three": 3, "four": 4, "five": 5}
    unique_users = set()
    cleaned_data = []

    for user in data:
        # Clean ratings
        raw_rating = str(user["rating"]).strip().lower()
        if(raw_rating in text_to_num):
            raw_rating = text_to_num[raw_rating]
        user["rating"] = raw_rating

        # Handle missing values
        raw_age = user.get("age")
        if(raw_age == None):
            user["age"] = None

        #de-duplicating of data
        if(user["name"] in unique_users):
            continue
        unique_users.add(user["name"])
        cleaned_data.append(user)

    return cleaned_data
```

```python
data = cleaning_data(data)
```

```python
# Get meaningful insights from the data

def get_insights(data):

    #Get Average rating
    total_ratings = 0
    for user in data:
        total_ratings += float(user["rating"])

    print(f"Average rating = {total_ratings/len(data)}")

    #Percentage of users with poor rating
    poor_ratings = 0
    for user in data:
        if(float(user["rating"]) < 3):
            poor_ratings += 1
    print(f"% of users with poor ratings = {poor_ratings/len(data) * 100}%")
```

```python
get_insights(data)
```
