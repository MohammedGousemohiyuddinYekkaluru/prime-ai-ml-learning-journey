```python
# Real data

# Oscar Winner Movies
oscar_movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]

years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 232] # in $M

# plot
plt.plot(years, oscar_revenue)

plt.title("Oscar MOvies Revenue in each year")
plt.xlabel("Years")
plt.ylabel("Revenue (in $M)")
```

```python
# Non-Oscar winner MOvies

non_oscar_movies = [
    "Slumdog Millionaire",
    "Avatar",
    "Inception",
    "Hugo",
    "Lincoln"
]
non_oscar_years = [2008, 2009, 2010, 2011, 2012]
non_oscar_revenue = [378, 2788, 829, 185, 275] #in $M

plt.plot(non_oscar_years, non_oscar_revenue, "o--r", linewidth=2, label="Non-Oscar Movies")
plt.plot(years, oscar_revenue, "<-y", label="Oscar-Movies")

plt.title("Oscar Movies vs. Non-Oscar Movies Revenue")
plt.xlabel("Years")
plt.ylabel("Revenue (in $M)")

plt.legend()
```

```python
# Bar charts

movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]

years = [2008, 2009, 2010, 2011, 2012]
revenue = [1005, 170, 427, 133, 232] # in $M

plt.bar(years, revenue, color="purple")
plt.title("Revenue in each year for Oscar Movies")
plt.xlabel("Years")
plt.ylabel("Revenue (in $M)")

for i in range(len(years)):
    plt.text(years[i], revenue[i] + 10, str(revenue[i]), ha="center")

plt.ylim(0, max(revenue) + 100)
plt.xlim(2007, max(years) + 2)
```

```python
# Compare 2 sets of Data
import numpy as np

oscar_movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]
oscar_revenue = [1005, 170, 427, 133, 232] # in $M

non_oscar_movies = [
    "Slumdog Millionaire",
    "Avatar",
    "Inception",
    "Hugo",
    "Lincoln"
]
non_oscar_revenue = [378, 2788, 829, 185, 275] #in $M

years = [2008, 2009, 2010, 2011, 2012]

x = np.arange(len(years))
width = 0.4

plt.bar(x-width/2, oscar_revenue, width, label="Oscar-Movies") #shifting to left
plt.bar(x+width/2, non_oscar_revenue, width, label="Non-Oscar Movies") # to right

plt.title("Oscar vs. Non-Oscar Movies Revenue")
plt.xlabel("Year")
plt.ylabel("Revenue (in $M)")
plt.legend()

plt.xticks(x, years)
```

```python
# Scatter Plots

people = ["Person A", "Person B", "Person C", "Person D", "Person E", 
          "Person F", "Person G", "Person H", "Person I", "Person J"]
age = [22, 25, 30, 35, 40, 45, 50, 55, 60, 65]
blood_pressure = [110, 115, 120, 122, 125, 130, 135, 123, 145, 150]

colors = ["green" if bp < 125 else "red" for bp in blood_pressure]
sizes = blood_pressure

plt.scatter(age, blood_pressure, s=sizes, cmap="OrRd", c=sizes)
plt.title("Age vs. BP")
plt.xlabel("Age")
plt.ylabel("Blood Pressure")

plt.grid()
plt.colorbar(label = "BP")

for i in range(len(age)):
    plt.annotate(people[i], xy=(age[i], blood_pressure[i]), xytext=(age[i]-6, blood_pressure[i]+2))
plt.xlim(15, max(age)+10)
plt.ylim(105, max(blood_pressure)+ 5)
```

```python
# Pie Charts

expenses = ["Salaries", "Rent", "Marketing", "R&D", "Miscellaneous"]
amounts = [500, 150, 120, 100, 50]

explode = [0, 0, 0, 0.2, 0]

plt.pie(amounts, labels=expenses, autopct="%1.1f%%",
        wedgeprops = {
            "edgecolor":"black",
            "linewidth": 1,
            "linestyle":"--",
            # "fill": False
        },
        explode = explode,
        startangle = 180,
        shadow = True
       )
```

```python
# Histogram

import numpy as np
import matplotlib.pyplot as plt

np.random.seed(10)
scores = np.random.normal(70, 10, 100)

plt.hist(scores, bins = 5, edgecolor = "black", color = "skyblue")
plt.title("Distribution of exam scores")
plt.xlabel("Marks")
plt.ylabel("Number of students")
```

```python
marks = [
    25, 38, 42, 55, 61, 67, 72, 78, 83, 88,
    91, 95, 47, 53, 59, 64, 69, 74, 79, 85,
    92, 97, 33, 45, 58, 62, 71, 76, 81, 89
]

# custom bins
bins = [0, 30, 70, 90, 100]
plt.hist(
    marks,
    bins=bins,
    edgecolor="black",
    color="lightblue"
)

plt.title("Distribution of Marks of 30 Students")
plt.xlabel("Marks")
plt.ylabel("Number of Students")
plt.grid(True)
```

```python
# axv line
marks = [
    25, 38, 42, 55, 61, 67, 72, 78, 83, 88,
    91, 95, 47, 53, 59, 64, 69, 74, 79, 85,
    92, 97, 33, 45, 58, 62, 71, 76, 81, 89
]
bins = [0, 30, 70, 90, 100]
plt.hist(
    marks,
    bins=bins,
    edgecolor="black",
    color="skyblue"
)

# line at passing marks - 33
plt.axvline(
    x=35,
    color="red",
    linestyle="--",
    linewidth=2,
    label="Pass Marks (35)"
)

plt.xlabel("Marks")
plt.ylabel("Number of Students")
plt.title("Distribution of Marks of 30 Students")
plt.grid(True)
plt.legend()
```

```python
# Multiple datasets
legit_transactions = [
    2.99, 5.49, 8.99, 12.50, 14.99, 19.99, 23.45, 29.99, 34.99, 39.50,
    45.00, 49.99, 55.25, 60.00, 75.99, 89.99, 120.50, 150.00, 199.99,
    249.99, 300.75, 450.00, 600.00, 850.00, 1200.00
]

fraud_transactions = [
    50, 100, 150, 200, 300, 500, 500, 750, 1000, 1000,
    1200, 1500, 1500, 2000, 2500, 3000, 3000
]

plt.hist(fraud_transactions, bins=10, label='Fraud', color='red', alpha=0.5, edgecolor="black")
plt.hist(legit_transactions, bins=10, label='Legit', color='green', alpha=0.5, edgecolor="black")

plt.xlabel('Transaction Amount ($)')
plt.ylabel('Frequency')
plt.title('Transaction Amount Distribution: Legitimate vs Fraudulent')
plt.legend()
plt.grid(True)
```
