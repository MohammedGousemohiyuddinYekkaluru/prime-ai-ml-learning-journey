```python
# Import the libraries
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the Tips Data
tips = sns.load_dataset('tips')
tips.head()
```

```python
# Pivot table of avg_tip
import pandas as pd
pivot_avg_tip = pd.pivot_table(
    tips,
    values="tip",
    index="day",
    columns="time",
    aggfunc="mean",
    observed=False
)
pivot_avg_tip
```

```python
# Heat map for pivot_avg_tip
sns.heatmap(
    pivot_avg_tip,
    cmap="coolwarm", annot=True
)
plt.title("Heatmap for average tip")
plt.tight_layout()
```
