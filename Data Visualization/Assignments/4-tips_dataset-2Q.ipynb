```python
# Import the libraries
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the Tips Data
tips = sns.load_dataset('tips')
tips.head()
```

```python
# Scatter plot --> total_bill vs. tip
sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="sex",
    style="smoker"
)

plt.title("total_bill vs. tip")
```

```python
# Do smokers tend to tip differently ?

# As i see in the plot Smokers appear to tip higher mainly because their total bills are higher.
# Smokers tend to tip higher than non-smokers even for similar bill amounts.
```
