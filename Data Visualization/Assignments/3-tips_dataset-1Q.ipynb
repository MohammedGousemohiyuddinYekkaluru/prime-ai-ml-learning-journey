```python
# Import the libraries
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the Tips Data
tips = sns.load_dataset('tips')
tips.head()
```

```python
# Box plot for total_bill on each day
sns.boxplot(
    data=tips,
    x="day",
    y="total_bill",
    hue="sex"
)

plt.grid()
```

```python
## Highest median bill
# On sunday Male has the highest median total bill
# On sat Female has the highest median total bill

## Comments on outliers
# Male customers show more outliers on Saturday and Sunday
# Female customers show outliers on Thursday and Saturday
```
