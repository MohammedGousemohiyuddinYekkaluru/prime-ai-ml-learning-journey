```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Dataset
np.random.seed(0)

scores = np.random.normal(70, 10, 200)

#histogram on scores
plt.hist(scores, edgecolor="black", color="skyblue", bins=5)

plt.grid()
plt.title("Distribution of Marks of 200 Students")
plt.xlabel("Scores")
plt.ylabel("Number of students")

#axv line for the mean
mean_score = scores.mean()
plt.axvline(
    x=mean_score,
    color="red",
    linestyle="-.",
    linewidth=2,
    label=f"AVG Marks ({mean_score:.2f})"
)

plt.legend()
```
